```python
data=corona_dataset21_csv.T
data
```

```python
data.columns
```

```python
data=data.drop('Date',axis=1)
```

```python
data=data.fillna(0)
data.info()
```

```python
lst=[]
for i in range(2538):
    if data['State'][i] not in lst[0::3]:
        sample_count=data['TotalSamples'][i]
        pos_count=data['Positive'][i]
        for j in range(i+1,2538):
            if data['State'][i]==data['State'][j]:
                sample_count+=data['TotalSamples'][j]
                pos_count+=data['Positive'][j]
        lst.append(data['State'][i])
        lst.append(sample_count)
        lst.append(pos_count)
print(lst)
```

```python
import pandas as pd
D=pd.DataFrame(data=lst[0::3],columns=['State'])
```

```python
D['Sample_count']=lst[1::3]
D['Positive_count']=lst[2::3]
D
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(25,20))
plt.xticks(rotation=80,horizontalalignment='center')
plt.bar(D.State,D.Positive_count,color=(0.2, 0.4, 0.6, 0.6))
plt.show()
```

```python
n=[]
for i in range(len(D)):
    n.append((D['Positive_count'][i]/D['Sample_count'][i])*100)
n=pd.DataFrame(n)
D['Infection_rate']=n
D
```

```python
data2=corona_dataset_csv.T 
data2
```

```python
data2=data2.drop('Date',axis=1)
data2
```

```python
data2=data2.fillna(0)
data2.info()
```

```python
lst1=[]
for k in range(3567):
    if data2['State/UnionTerritory'][k] not in lst1[0::4]:
        Cured=data2['Cured'][k]
        Deaths=data2['Deaths'][k]
        Confirmed=data2['Confirmed'][k]
        for l in range(k+1,3567):
            if data2['State/UnionTerritory'][k]==data2['State/UnionTerritory'][l]:
                Cured+=data2['Cured'][l]
                Deaths+=data2['Deaths'][l]
                Confirmed+=data2['Confirmed'][l]
        lst1.append(data2['State/UnionTerritory'][k])
        lst1.append(Cured)
        lst1.append(Deaths)
        lst1.append(Confirmed)
print(lst1)
```

```python
D1=pd.DataFrame(lst1[0::4],columns=['State/UnionTerritory'])
```

```python
D1.rename(columns={'State/UnionTerritory':'States'},inplace=True)
```

```python
D1['Cured']=lst1[1::4]
D1['Deaths']=lst1[2::4]
D1['Confirmed']=lst1[3::4]
D1
```

```python
c=[]
d=[]
for i in range(len(D1)):
    c.append((D1['Cured'][i]/D1['Confirmed'][i])*100)
    d.append((D1['Deaths'][i]/D1['Confirmed'][i])*100)
c=pd.DataFrame(c)
d=pd.DataFrame(d)
D1['Recovery_Rate']=c
D1['Death_Rate']=d
D1
```

```python
from matplotlib.pyplot import figure
figure(figsize=(25,20))
plt.xticks(rotation=80,horizontalalignment='center')
plt.bar(D1.States,D1.Recovery_Rate,color=(0.2, 0.4, 0.6, 0.6),label='Recovery Rate')
plt.bar(D1.States,D1.Death_Rate,color=(0.2,0.2,0.6,0.6),label='Death Rate')
plt.show()
```
